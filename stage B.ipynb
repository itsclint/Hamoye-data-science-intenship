```python
# Normalize dataset using a MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(df),
                             columns=df.columns)

features_df = normalised_df.drop(columns=['Appliances'])
target = normalised_df['Appliances']


# dataset is split 70:30, training set and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_df, target, test_size=0.3, random_state=42)
```

```python
#import Liner model 
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()

#fit the model to the training dataset
linear_model.fit(x_train, y_train)

#obtain predictions
predicted_values = linear_model.predict(x_test)
```

```python
print(f'Coefficients: {linear_model.coef_}')
print(f'Intercept: {linear_model.intercept_}')
print(f'R^2 score: {linear_model.score(x_test, y_test)}')
```

```python
#comparing the effects of regularisation
def get_weights_df(model, feat, col_name):
  #this function returns the weight of every feature
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)
```

```python
linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')

final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
final_weights
```

```python
def linearity_test(model, y):
    ''' 
    this function visualizes the assumption of linearity in the model.
    it plots a graph of the observed vs the predicted values,
    and a graph of the residuals vs the predicted values.
    
    Args:
    * model - fitted OLS model from statsmodels.api
    * y - observed values

    '''
    
    fitted_vals = model.predict()
    resids = model.resid
    
    fig, ax = plt.subplots(1,2)
    
    sns.regplot(x=fitted_vals, y=y, lowess=True, ax=ax[0], line_kws={'color': 'red'})
    ax[0].set_title('Observed vs. Predicted Values', fontsize=16)
    ax[0].set(xlabel='Predicted', ylabel='Observed')

    sns.regplot(x=fitted_vals, y=resids, lowess=True, ax=ax[1], line_kws={'color': 'red'})
    ax[1].set_title('Residuals vs. Predicted Values', fontsize=16)
    ax[1].set(xlabel='Predicted', ylabel='Residuals')
    
```

```python
import statsmodels.api as sm

X_constant = sm.add_constant(x_test)
lin_reg = sm.OLS(y_test, X_constant).fit()
lin_reg.summary()
```

```python
linearity_test(lin_reg, y_test)
```
